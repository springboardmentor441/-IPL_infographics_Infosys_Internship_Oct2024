NULL VALUES

identifying null values

```python
print(match_data.isnull().sum()) # count null values in each column
```

Null values way 1: fill it with some value

```python
# so null values are there in city column so we will replace null values with the value of a city say Chandigarh
match_data['city']=match_data['city'].fillna('Chandigarh')
```

```python
print(match_data.isnull().sum())
# now we can see there are no null values in the city column
```

way 2: fill it with median/mean

```python
# now see the result margin column has null values and it's numerical values so we can replace them with median/mean/mode
# with mean
match_data['result_margin']=match_data['result_margin'].fillna(match_data['result_margin'].mean())
```

```python
match_data.isnull().sum()
```

```python
# now see target runs also has null values and it's numerical value, so we can replace with the median/mean/mode value.
# with median
match_data['target_runs']=match_data['target_runs'].fillna(match_data['target_runs'].median())
```

```python
match_data.isnull().sum()
```

```python
# now see the column target overs it's numerical value
match_data['target_overs']=match_data['target_overs'].fillna(match_data['target_overs'].mode()[0])
```

The mode() function in Pandas returns the most frequently occurring value(s) in a Series or DataFrame column.
If there is more than one mode (multiple values occur with the same highest frequency), it returns all of them as a Series.

```python
match_data.isnull().sum()
```

```python
# can use forward and backward fill for the text values
match_data['player_of_match']=match_data['player_of_match'].fillna(method='ffill')
match_data['winner']=match_data['winner'].fillna(method='bfill')
```

```python
match_data.isnull().sum()
```

```python
match_data['method']=match_data['method'].fillna(0)
```

```python
match_data.isnull().sum()
```

```python
print(match_data['method'].dtypes)
```
