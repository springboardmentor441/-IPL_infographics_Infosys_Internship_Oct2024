```python
import pandas as pd
```

```python
matches_data = pd.read_csv('matches.csv')
```

```python
# Display the first few rows
print(matches_data.head())

# Check available columns
print(matches_data.columns)
```

```python
# Verify the data in the 'city' column
print(matches_data['city'].unique())
```

```python
# Count the number of matches in each city
city_match_counts = matches_data['city'].value_counts()
```

```python
# Get the top 20 cities
top_20_cities = city_match_counts.head(20)

# Convert to a DataFrame 
top_20_cities_df = top_20_cities.reset_index()
top_20_cities_df.columns = ['City', 'Match Count']
```

```python
print(top_20_cities_df)
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_20_cities_df['City'], top_20_cities_df['Match Count'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top 20 Cities by Number of Matches Played')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```
