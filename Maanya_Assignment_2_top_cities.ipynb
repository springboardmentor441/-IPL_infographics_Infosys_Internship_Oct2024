```python
import pandas as pd
```

```python
match_data=pd.read_csv("matches.csv")
```

```python
match_data
```

To find the top cities, we will group the cities column with the match id for each city and then we will calculate the number of matches played in that city. And more the number of matches played, the city is the top city.

```python
match_data['city'].unique()
```

```python
match_data.replace("Bengaluru","Banglore",inplace=True)
match_data.replace("Navi Mumbai","Mumbai",inplace=True)
match_data.replace("Abu Dhabi","Dubai",inplace=True)
match_data.replace("Mohali","Chandigarh",inplace=True)
```

```python
match_data
```

```python
city_match_count=match_data.groupby('city')['id'].nunique().reset_index(name='match_count')
```

```python
city_match_count
```

```python
top_cities=city_match_count.sort_values(by='match_count',ascending=False).reset_index(drop=True)
```

```python
print(top_cities)
```
