## Combining

```python
# Create a second DataFrame with a subset of rows or modified data
matches_extra = Matches.sample(frac=0.5, random_state=1)  # Using a random sample of 50% of rows
```

```python
# Concatenate the original data and the new sample
combined_matches = pd.concat([Matches, matches_extra], ignore_index=True)
```

```python
# Set multi-indexes for joining
left = Matches.set_index(['team1', 'season'])
right = matches_extra.set_index(['team1', 'season'])

# Join with suffixes to differentiate overlapping columns
joined_matches = left.join(right, lsuffix='_orig', rsuffix='_extra')
```
