```python
# Runs by players
player_match_stats = df.groupby(['batter', 'match_id'])['batsman_runs'].sum().reset_index()

player_stats = player_match_stats.groupby('batter').agg(
    mean_runs=('batsman_runs', 'mean'),
    median_runs=('batsman_runs', 'median'),
    mode_runs=('batsman_runs', lambda x: x.mode()[0] if not x.mode().empty else None),
    total_runs=('batsman_runs', 'sum')
)
player_stats = player_stats.sort_values(by='total_runs', ascending=False)
player_stats.head(20)
```

```python
# batting Averages of players

total_runs = df.groupby('batter')['batsman_runs'].sum()
times_out = df[df['is_wicket'] == 1].groupby('batter')['is_wicket'].count()

batsman_stats = pd.DataFrame({'total_runs': total_runs, 'times_out': times_out})

# Calculating the batting average
batsman_stats['batting_average'] = batsman_stats['total_runs'] / batsman_stats['times_out']

batsman_stats = batsman_stats.sort_values(by='total_runs', ascending=False)
batsman_stats.head(20)
```

```python
# Calculating total balls faced by each batter
balls_faced = df.groupby('batter').size()

# Calculating total runs, times out, and balls faced
total_runs = df.groupby('batter')['batsman_runs'].sum()
times_out = df[df['is_wicket'] == 1].groupby('batter')['is_wicket'].count()

batsman_stats = pd.DataFrame({
    'total_runs': total_runs,
    'times_out': times_out,
    'balls_faced': balls_faced
})

# Calculating batting average
batsman_stats['batting_average'] = batsman_stats['total_runs'] / batsman_stats['times_out']

# Calculating strike rate
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced']) * 100

# Calculating total 100s and 50s
player_match_stats = df.groupby(['batter', 'match_id'])['batsman_runs'].sum().reset_index()
player_match_stats['hundreds'] = player_match_stats['batsman_runs'].apply(lambda x: 1 if x >= 100 else 0)
player_match_stats['fifties'] = player_match_stats['batsman_runs'].apply(lambda x: 1 if 50 <= x < 100 else 0)

hundreds = player_match_stats.groupby('batter')['hundreds'].sum()
fifties = player_match_stats.groupby('batter')['fifties'].sum()


batsman_stats['total_hundreds'] = hundreds
batsman_stats['total_fifties'] = fifties
batsman_stats.fillna({'total_hundreds': 0, 'total_fifties': 0}, inplace=True)

batsman_stats = batsman_stats.sort_values(by='total_runs', ascending=False)
batsman_stats.head(20)
```

## Top 10 best batsman

```python
# finding the best batsman
batsman_stats['composite_score'] = (
    (batsman_stats['total_runs'] / batsman_stats['total_runs'].max()) * 0.25 + 
    (batsman_stats['batting_average'] / batsman_stats['batting_average'].max()) * 0.20 + 
    (batsman_stats['strike_rate'] / batsman_stats['strike_rate'].max()) * 0.20 + 
    (batsman_stats['total_hundreds'] / batsman_stats['total_hundreds'].max()) * 0.15 + 
    (batsman_stats['total_fifties'] / batsman_stats['total_fifties'].max()) * 0.10 +
    ((1 / batsman_stats['times_out']) / (1 / batsman_stats['times_out']).max()) * 0.10
)

# Sorting players by composite score
top_10_players = batsman_stats.sort_values(by='composite_score', ascending=False).head(10)

# Displaying the top 10 players as a table
top_10_players[['total_runs', 'batting_average', 'strike_rate', 'total_hundreds', 'total_fifties', 'balls_faced', 'composite_score']]
```

# Design Document: Determining the Best Batter

## 1. Objective
To design a strategy for evaluating and determining the best batter based on performance metrics derived from IPL data. The evaluation considers statistical performance, consistency, and impact on matches.

## 2. Strategy for Evaluation
The strategy involves analyzing batting performance metrics across various dimensions and assigning weights to these parameters to calculate a composite score for each batter. The batter with the highest score will be deemed the best.

## 3. Features (Parameters) to Be Considered
The following parameters will be analyzed to evaluate batting performance:

- **Total Runs Scored (total_runs):**
  - *Significance:* Indicates overall contribution to the teamâ€™s performance.

- **Batting Average (batting_average):**
  - *Formula:* `total_runs / times_out`
  - *Significance:* Reflects consistency and reliability.

- **Strike Rate (strike_rate):**
  - *Formula:* `(total_runs / balls_faced) * 100`
  - *Significance:* Measures scoring efficiency and impact in limited-over formats.

- **Number of Hundreds (total_hundreds):**
  - *Significance:* Highlights match-defining performances.

- **Number of Fifties (total_fifties):**
  - *Significance:* Indicates ability to provide substantial contributions.

- **Balls Faced (balls_faced):**
  - *Significance:* Helps balance strike rate and contribution over time.

- **Consistency Score (consistency):**
  - *Formula:* Standard deviation of runs across matches.
  - *Significance:* Reflects a player's ability to maintain consistent performance.

## 4. Weight Metrics for Different Parameters
| Parameter         | Weight |
|--------------------|--------|
| Total Runs         | 25%    |
| Batting Average    | 20%    |
| Strike Rate        | 20%    |
| Number of Hundreds | 15%    |
| Number of Fifties  | 10%    |
| Consistency Score  | 10%    |

## 5. Composite Score Calculation
The composite score for each batter is calculated using a weighted sum of normalized values for each parameter.

## 6. Implementation Steps
1. **Data Cleaning and Preprocessing:**
   - Handle missing values.
   - Ensure data consistency (e.g., handling invalid or duplicate entries).

2. **Feature Engineering:**
   - Derive features such as batting average, strike rate, consistency score, and match-level stats.

3. **Normalization:**
   - Normalize each parameter to ensure comparability (e.g., Min-Max scaling).

4. **Weight Application:**
   - Apply weights to calculate the composite score.

5. **Ranking:**
   - Rank batters based on their composite scores.

## Overall team batting results

```python
# Runs by teams

team_match_stats = df.groupby(['batting_team', 'match_id'])['total_runs'].sum().reset_index()

team_stats = team_match_stats.groupby('batting_team').agg(
    mean_runs=('total_runs', 'mean'),
    median_runs=('total_runs', 'median'),
    mode_runs=('total_runs', lambda x: x.mode()[0] if not x.mode().empty else None),
    total_runs=('total_runs', 'sum')
)
team_stats = team_stats.sort_values(by='total_runs', ascending=False)
team_stats
```

```python
# Wickets by players
wickets_data = df[df['is_wicket'] == 1]

bowler_match_wickets = wickets_data.groupby(['bowler', 'match_id'])['is_wicket'].sum().reset_index()

bowler_stats = bowler_match_wickets.groupby('bowler').agg(
    mean_wickets=('is_wicket', 'mean'),
    median_wickets=('is_wicket', 'median'),
    mode_wickets=('is_wicket', lambda x: x.mode()[0] if not x.mode().empty else None),
    total_wickets=('is_wicket', 'sum')
)
bowler_stats = bowler_stats.sort_values(by='total_wickets', ascending=False)

bowler_stats.head(20)
```

```python
# Calculating total wickets taken by each bowler
total_wickets = df[df['is_wicket'] == 1].groupby('bowler')['is_wicket'].count()

# Calculating total runs conceded by each bowler
runs_conceded = df.groupby('bowler')['total_runs'].sum()

# Calculating total balls bowled by each bowler
balls_bowled = df[df['extra_runs'] == 0].groupby('bowler').size()

# Calculating economy rate
economy_rate = runs_conceded / (balls_bowled / 6)

# Calculating bowling average
bowling_average = runs_conceded / total_wickets

# Calculating strike rate
bowling_strike_rate = balls_bowled / total_wickets


bowler_stats = pd.DataFrame({
    'total_wickets': total_wickets,
    'runs_conceded': runs_conceded,
    'balls_bowled': balls_bowled,
    'economy_rate': economy_rate,
    'bowling_average': bowling_average,
    'bowling_strike_rate': bowling_strike_rate
})
bowler_stats = bowler_stats.sort_values(by='total_wickets', ascending=False)
bowler_stats.head(20)
```

## Top 10 best bowlers

```python
# Finding the best bowlers
bowler_stats['composite_score'] = (
    (bowler_stats['total_wickets'] / bowler_stats['total_wickets'].max()) * 0.30 + 
    ((1 / bowler_stats['economy_rate']) / (1 / bowler_stats['economy_rate']).max()) * 0.25 + 
    ((1 / bowler_stats['bowling_average']) / (1 / bowler_stats['bowling_average']).max()) * 0.20 + 
    ((1 / bowler_stats['bowling_strike_rate']) / (1 / bowler_stats['bowling_strike_rate']).max()) * 0.25
)

top_10_bowlers = bowler_stats.sort_values(by='composite_score', ascending=False).head(10)
top_10_bowlers[['total_wickets', 'economy_rate', 'bowling_average', 'bowling_strike_rate', 'composite_score']]
```

# Design Document: Determining the Best Bowler

## 1. Objective
To design a strategy for evaluating and determining the best bowler based on performance metrics derived from IPL data. The evaluation considers efficiency, consistency, and match impact.

## 2. Strategy for Evaluation
The strategy involves analyzing bowling performance metrics and assigning weights to these parameters to compute a composite score for each bowler. The bowler with the highest composite score will be identified as the best.

## 3. Features (Parameters) to Be Considered
The following parameters will be used for evaluation:

- **Total Wickets Taken (total_wickets):**
  - *Significance:* Indicates the bowlerâ€™s effectiveness in taking wickets.

- **Economy Rate (economy_rate):**
  - *Formula:* `runs_conceded / (balls_bowled / 6)`
  - *Significance:* Measures the bowler's efficiency in containing runs.

- **Bowling Average (bowling_average):**
  - *Formula:* `runs_conceded / total_wickets`
  - *Significance:* Reflects the cost of taking wickets.

- **Bowling Strike Rate (bowling_strike_rate):**
  - *Formula:* `balls_bowled / total_wickets`
  - *Significance:* Measures how frequently the bowler takes wickets.

- **Balls Bowled (balls_bowled):**
  - *Significance:* Highlights the workload and reliability.

## 4. Weight Metrics for Different Parameters
| Parameter              | Weight |
|-------------------------|--------|
| Total Wickets          | 30%    |
| Economy Rate           | 25%    |
| Bowling Average        | 20%    |
| Bowling Strike Rate    | 25%    |

## 5. Composite Score Calculation
The composite score for each bowler will be calculated using the weighted sum of normalized values for each parameter.

### Normalization
Each parameter will be normalized using Min-Max scaling or inverse scaling for parameters where lower values are better (e.g., economy rate).

   - Handle missing or invalid values.
   - Ensure accurate filtering of bowling-related metrics (e.g., excluding extras for balls bowled).

2. **Feature Engineering:**
   - Derive features like economy rate, bowling average, and strike rate.

3. **Normalization and Weighting:**
   - Normalize metrics and apply weights to compute the composite score.

4. **Ranking:**
