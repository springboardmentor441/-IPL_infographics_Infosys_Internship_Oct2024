**RUNS ANALYSIS BY PLAYER AND TEAM**

```
import pandas as pd
```

```
results=pd.read_csv('deliveries.csv')
```

```
results.head()
```

```
filtered_data = results[['match_id', 'batter', 'batsman_runs']]
filtered_data
```

```
player_scores = filtered_data.groupby(['match_id', 'batter']).sum().reset_index()
player_scores
```

```
overall_scores = player_scores.groupby('batter')['batsman_runs'].agg(['mean', 'median']).reset_index()
overall_scores
```

```
modes = player_scores.groupby('batter')['batsman_runs'].agg(
    lambda x: (
        "No Mode" if x.nunique() == len(x) else  # No mode (all unique values)
        x.mode().tolist() if len(x.mode()) > 1 else  # Multiple modes
        x.mode()[0]  # Single mode
    )
)
modes = modes.reset_index(name='mode')
modes
```

```
overall_stats = overall_scores.merge(modes, on='batter')
overall_stats
```

```
specific_batter=player_scores[player_scores['batter']=='BB McCullum']
specific_batter
```

```
individual_stats =overall_stats[overall_stats['batter'] == specific_batter['batter'].iloc[0]]
individual_stats
```

```
filtered_data = results[['match_id', 'batting_team', 'total_runs']]
filtered_data
```

```
team_scores = filtered_data.groupby(['match_id', 'batting_team']).sum().reset_index()
team_scores
```

```
specific_team=team_scores[team_scores['batting_team']=='Kochi Tuskers Kerala']
specific_team
```

```
overall_team_scores = team_scores.groupby('batting_team')['total_runs'].agg(['mean', 'median']).reset_index()
overall_team_scores
```

```
team_modes = team_scores.groupby('batting_team')['total_runs'].agg(
    lambda x: (
        "No Mode" if x.nunique() == len(x) else  # No mode (all unique values)
        x.mode().tolist() if len(x.mode()) > 1 else  # Multiple modes
        x.mode()[0]  # Single mode
    )
)
team_modes = team_modes.reset_index(name='mode')
team_modes
```

```
overall_team_stats = overall_team_scores.merge(team_modes, on='batting_team')
overall_team_stats
```

```
batting_average = overall_stats.sort_values(by='mean', ascending=False).head(10)
batting_average
```

```
highest_strike_rate = overall_stats.sort_values(by='median', ascending=False).head(10)
highest_strike_rate
```

```
team_average = overall_team_stats.sort_values(by='mean', ascending=False).head(10)
team_average
```

```
highest_team_average = overall_team_stats.sort_values(by='median', ascending=False).head(10)
highest_team_average
```

```
player_total_runs_in_season=player_scores.groupby('batter')['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs',ascending=False).head(10)
player_total_runs_in_season
```

```
team_total_runs_in_season=team_scores.groupby('batting_team')['total_runs'].sum().reset_index().sort_values(by='total_runs',ascending=False).head(10)
team_total_runs_in_season
```

Design Document: Determining the Best Batter
Objective

To design a comprehensive methodology to evaluate and rank batters based on their performance using a combination of statistical and contextual parameters.

Strategy for Evaluation
Multifactor Analysis: Assess batters on multiple metrics like runs scored, strike rate, consistency, and impact in key situations.

Weighting System: Assign weights to each metric based on its relative importance to the batter’s performance and the game context.

Normalization: Normalize data to ensure comparability across different formats (T20, ODI, Test) and eras.

Final Scoring: Combine weighted metrics into a composite score to rank batters.
