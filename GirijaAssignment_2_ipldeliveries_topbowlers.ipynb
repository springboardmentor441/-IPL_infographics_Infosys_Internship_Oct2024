statistics for runs

```python
player_runs = deliveries.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
player_runs= deliveries.groupby('batter')['batsman_runs'].sum().reset_index()
player_runs.columns=["batter","total_runs"]
player_runs
```

```python
batter_knockout=deliveries.groupby("batter")["is_wicket"].sum()
batter_info=pd.merge(player_runs,batter_knockout,on="batter",how="left")
batter_info.columns=["batter","total_runs","total_times_out"]
batter_knockout.head(5)
```

```python
total_matches=deliveries.groupby("batter")["match_id"].nunique().reset_index()
total_matches.columns=["batter","total_matches_played"]
batter_stats=pd.merge(batter_info,total_matches,on="batter",how="left")
```

```python
batter_stats
```

```python
batter_stats["batting_average"]=batter_stats["total_runs"]/batter_stats["total_times_out"]
```

```python
batter_stats["batting_average"].replace(np.inf,np.nan,inplace=True)
batter_stats.loc[batter_stats["total_times_out"]==0,"batting_average"]=np.nan
top_5_batter_average=batter_stats.sort_values(by="batting_average",ascending=False)
top_5_batter_average.head(5)
```

```python
batter_stats["player_score"]=(batter_stats["batting_average"]+(batter_stats["total_runs"]/1000)+(batter_stats["total_matches_played"]/10))
best_player=batter_stats.sort_values(by="player_score",ascending=False)
best_player.head(5)
```

```python
batter_stats = deliveries.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    balls_faced=('ball', 'count'),
    dismissals=('is_wicket', 'sum'),
    centuries=('batsman_runs', lambda x: (x >= 100).sum()),
    fifties=('batsman_runs', lambda x: ((x >= 50) & (x < 100)).sum())
).reset_index()

# Calculating batting average and strike rate
batter_stats['batting_average'] = batter_stats['total_runs'] / batter_stats['dismissals']
batter_stats['strike_rate'] = (batter_stats['total_runs'] / batter_stats['balls_faced']) * 100

# Replacing  NaN averages for not-out players with their total runs
batter_stats['batting_average'].fillna(batter_stats['total_runs'], inplace=True)

# :Normalize metrics
batter_stats['normalized_average'] = (batter_stats['batting_average'] / batter_stats['batting_average'].max()) * 100
batter_stats['normalized_strike_rate'] = (batter_stats['strike_rate'] / batter_stats['strike_rate'].max()) * 100
batter_stats['normalized_centuries'] = (batter_stats['centuries'] / batter_stats['centuries'].max()) * 100
batter_stats['normalized_fifties'] = (batter_stats['fifties'] / batter_stats['fifties'].max()) * 100

# Handle cases with no centuries or fifties (avoid NaN in normalization)
batter_stats['normalized_centuries'].fillna(0, inplace=True)
batter_stats['normalized_fifties'].fillna(0, inplace=True)

#Compute overall score
w1, w2, w3, w4 = 0.3, 0.4, 0.15, 0.05
batter_stats['overall_score'] = (
    batter_stats['normalized_average'] * w1 +
    batter_stats['normalized_strike_rate'] * w2 +
    batter_stats['normalized_centuries'] * w3 +
    batter_stats['normalized_fifties'] * w4
)

# Find the best batter
best_batter = batter_stats.sort_values(by='overall_score', ascending=False).head(5)

# Display results
print("Best Batter Based on Criteria:")
print(best_batter)
```
