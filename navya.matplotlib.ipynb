```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the IPL dataset
file_path = 'matches.csv'  
ipl_data = pd.read_csv("matches.csv")

# Checking the first few rows to understand the data structure 
print(ipl_data.head())

# Counting  the number of matches played at each venue
venue_match_counts = ipl_data['venue'].value_counts()

# Plotting a bar chart for the number of matches played at each venue
plt.figure(figsize=(12, 8))  
sns.barplot(x=venue_match_counts.values, y=venue_match_counts.index, palette='viridis')

# Add titles and x and y labels with font sizes
plt.title('Number of Matches Played at Each Venue', fontsize=16)
plt.xlabel('Number of Matches', fontsize=12)
plt.ylabel('Venue', fontsize=12)

# Optimize layout and show the plot
plt.tight_layout()
plt.show()
```
