```python
import numpy as np # any numerical operations
import pandas as pd # handling datasets
import matplotlib.pyplot as plt # creating static, animated, and interactive visualizations
import seaborn as sns #a high-level interface for drawing attractive and informative statistical graphics
```

```python
matches=pd.read_csv('/content/matches.csv')
```

```python
# Aim: To count the number of matches played at each venue.

venue_matches = matches['venue'].value_counts()
venue_matches
```

```python
len(matches['venue'].unique())
```

```python
# Aim: To create a bar chart to represent the number of matches per venue.

plt.figure(figsize=(14,7))
venue_matches.plot(kind='bar')
plt.title('Number of matches per venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.show()
```

```python
# Aim: To determine the most popular venues.

# Let's get the top 5 venues.
top_5_venues = venue_matches.head(5)
print("Top 5 Venues:\n", top_5_venues)
```

```python
# Aim: To analyze match outcomes (wins, losses, etc.)

# Count the number of wins for each team.
team_wins = matches['winner'].value_counts()
print("\nTeam Wins:\n", team_wins)
```

```python
# Visualize match outcomes
plt.figure(figsize=(14,7))
team_wins.head(10).plot(kind='bar') # visualize top 10 teams for better readability
plt.title('Number of wins per team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
# Analyze toss decisions
toss_decisions = matches['toss_decision'].value_counts()
print("\nToss Decisions:\n", toss_decisions)
```

```python
plt.figure(figsize=(6,6))
plt.pie(toss_decisions, labels=toss_decisions.index, autopct='%1.1f%%', startangle=90)
plt.title("Toss Decision Distribution")
plt.show()
```

```python
# Combined Analysis (Example: Wins based on Toss Decision)
# This is a more complex analysis and provides deeper insights.

# For simplicity let's only look at the top 5 teams
top_teams = team_wins.head(5).index

# Create a subset of data only for the top teams
matches_top_teams = matches[matches['winner'].isin(top_teams)]
```

```python
pd.crosstab(matches_top_teams['winner'],matches_top_teams['toss_decision']).plot(kind='bar', figsize=(12,6))
plt.title("Wins based on Toss Decision (Top 5 teams)")
plt.xlabel("Winning Team")
plt.ylabel("Count")
plt.show()
```

```python
# creating an interactive map showing venue locations using Folium with sample dataset
#  that has various venues(groundName),its latitude and longitude.

# Load the dataset
venues=pd.read_csv('/content/ground_location.csv')
```

```python
venues
```

```python
#importing folium package

import folium
```

```python
# Create a map centered around India
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
```

```python
# Add markers for each venue
for index, row in venues.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['GroundName'],  # Show venue name in popup
    ).add_to(m)
```
