## **Match Distribution by Season**

```python
matches_per_season = matches['season'].value_counts().sort_index()
matches_per_season.plot(kind='bar', color='skyblue', title="Match Distribution by Season")
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.show()
```

```python
top_players = matches['player_of_match'].value_counts().head(10)
top_players.plot(kind='bar', color='orange', title="Top Players of the Match")
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.show()
```

```python
matches_per_city = matches['city'].value_counts()
matches_per_city.plot(kind='bar', color='purple', title="Match Locations")
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```

```python
wins_per_team = matches['winner'].value_counts()
wins_per_team.plot(kind='bar', color='green', title="Teams with the Most Wins")
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
toss_decision = matches.groupby(['season', 'toss_decision']).size().unstack()
toss_decision.plot(kind='bar', stacked=True, title="Toss Decision Analysis")
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(matches['result_margin'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title("Distribution of Victory Margin")
plt.xlabel("Result Margin (runs/wickets)")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_target_runs = matches.groupby('season')['target_runs'].mean()
avg_target_runs.plot(kind='line', marker='o', color='red', title="Average Target Runs by Season")
plt.xlabel('Season')
plt.ylabel('Average Target Runs')
plt.show()
```

```python
toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]
toss_win_outcome = toss_win_match_win['toss_decision'].value_counts()
toss_win_outcome.plot(kind='bar', color='teal', title="Impact of Toss Decision on Match Outcome")
plt.xlabel('Toss Decision')
plt.ylabel('Wins after Winning Toss')
plt.show()
```
