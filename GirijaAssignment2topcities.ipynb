**FINDING TOP MOST 20 CITIES**

Null values are replaced with venue.Duplicate values are replaced. Matches happened in each city is calculated using value_counts() and stored in topcities. To get top 20 cities head(20) is used with topcities and stored in top_20_cities. And the 20 cities with its values is plotted using matplotlib


```python
matches=pd.read_csv("/content/drive/MyDrive/Copy of matches.csv")
```

```python
matches.info()
```

```python
matches.sample(15)
```

```python
matches.city.isnull().sum()
```

```python
matches.city.fillna(matches.venue,inplace=True)
```

```python
matches.city.isnull().sum()
```

```python
matches.city.unique()
```

```python
matches.city.nunique()
```

```python
matches.city.replace("Dubai International Cricket Stadium","Dubai",inplace=True)
matches.city.replace("Sharjah Cricket Stadium","Sharjah",inplace=True)
```

```python
matches.replace("Bengaluru","Bangalore",inplace=True)
```

```python
matches.city.unique()
```

```python
matches.city.nunique()
```

```python
topcities=matches.city.value_counts()
top_20_cities=topcities.head(20)
top_20_cities
```

```python
top_20_cities.plot(kind='bar',color='darkviolet')
plt.title("Top 20 cities based on number of matches",color='green')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```
