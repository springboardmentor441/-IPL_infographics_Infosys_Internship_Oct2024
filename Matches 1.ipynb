# Maps

```python
###A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values
Matches_target_overs_mean = Matches.target_overs.mean()
Matches.target_overs.map(lambda p: p - Matches_target_overs_mean)
```

```python
# Define the function to adjust 'result_margin' in each row
def remean_result_margin(row):
    # Subtract the average 'result_margin' from each row's 'result_margin'
    row['result_margin'] = row['result_margin'] - Matches['result_margin'].mean()
    return row

# Apply the function to each row
Matches = Matches.apply(remean_result_margin,axis='columns')
```

```python
import matplotlib.pyplot as plt
Matches['date'] = pd.to_datetime(Matches['date'], errors='coerce')

# Sort data by date to ensure a proper timeline
Matches = Matches.sort_values(by='date')

# Plotting the result margin over time
plt.figure(figsize=(14, 6))
plt.plot(Matches['date'], Matches['result_margin'], color='blue', alpha=0.7, label='Result Margin')
plt.title('Trend of Result Margin Over Time')
plt.xlabel('Date')
plt.ylabel('Result Margin')
plt.legend()
plt.show()
```

```python
# Convert 'date' column to datetime format
Matches['date'] = pd.to_datetime(Matches['date'], errors='coerce')

# Sort by date
Matches = Matches.sort_values(by='date').reset_index(drop=True)

# Calculate the change in 'result_margin' between consecutive matches
Matches['margin_change'] = Matches['result_margin'].diff()

# Identify up and down trends
Matches['trend'] = Matches['margin_change'].apply(lambda x: 'uprun' if x > 0 else ('downrun' if x < 0 else 'no change'))

# Plotting
plt.figure(figsize=(14, 6))

# Plot uprun segments
plt.plot(Matches['date'][Matches['trend'] == 'uprun'], 
         Matches['result_margin'][Matches['trend'] == 'uprun'], 
         color='green', label='Uprun (Increase)', alpha=0.7)

# Plot downrun segments
plt.plot(Matches['date'][Matches['trend'] == 'downrun'], 
         Matches['result_margin'][Matches['trend'] == 'downrun'], 
         color='red', label='Downrun (Decrease)', alpha=0.7)

# Add title and labels
plt.title('Uprun and Downrun Trends in Result Margin Over Time')
plt.xlabel('Date')
plt.ylabel('Result Margin')
plt.legend()
plt.show()
```
