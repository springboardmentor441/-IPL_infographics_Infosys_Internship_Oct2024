```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matches_path='/content/matches.csv'
```

```python
df = pd.read_csv(matches_path)
```

```python
df.head()
```

```python
city_match_count = df['city'].value_counts()
```

```python
top_20_cities = city_match_count.head(20)
print(top_20_cities)
```

```python
# Plot the top 20 cities in a pie chart
plt.figure(figsize=(30, 15))
top_20_cities.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Top 20 Cities with Most IPL Matches Played')
plt.ylabel('')  # Hides the y-axis label as it's not needed in a pie chart
plt.show()
```
