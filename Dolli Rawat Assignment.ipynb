```python
import pandas as pd
```

```python
Matches = pd.read_csv("matches.csv")
```

```python
Matches = pd.read_csv("matches.csv", index_col=0)
Matches.head()
```

```python
# Step 1: Fill missing values
# Fill 'city' with 'venue' if 'city' is missing
Matches['city'] = Matches['city'].fillna(Matches['venue'])
```

```python
# Fill 'winner' and 'player_of_match' with placeholder text if missing
Matches['winner'] = Matches['winner'].fillna('No result')
Matches['player_of_match'] = Matches['player_of_match'].fillna('No award')
```

```python
# Fill missing numeric columns with 0 or appropriate values
Matches['result_margin'] = Matches['result_margin'].fillna(0)
Matches['target_runs'] = Matches['target_runs'].fillna(0)
Matches['target_overs'] = Matches['target_overs'].fillna(0)
```

```python
# Fill missing umpire names with 'unknown'
Matches['umpire1'] = Matches['umpire1'].fillna('unknown')
Matches['umpire2'] = Matches['umpire2'].fillna('unknown')
```

```python
# Step 2: Drop duplicates (if any)
Matches = Matches.drop_duplicates()
```

```python
# Step 5: Check unique values in each column
for column in Matches.columns:
    unique_values = Matches[column].nunique()
    print(f"{column}: {unique_values} unique values")
```

```python
# Display a summary of the cleaned data
print(Matches.info())
print(Matches.head())
```
