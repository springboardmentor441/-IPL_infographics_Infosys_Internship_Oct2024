```python
#calculate the mean, median, and mode for runs

from scipy import stats

# Calculate the mean
mean_runs = df['total_runs'].mean()
print(f"Mean runs: {mean_runs}")

# Calculate the median
median_runs = df['total_runs'].median()
print(f"Median runs: {median_runs}")

# Calculate the mode
mode_runs = stats.mode(df['total_runs'])
print(f"Mode runs: {mode_runs}")
```

```python
#calculate the mean, median, and mode for wickets

from scipy import stats

# Calculate the mean
mean_wickets = df['is_wicket'].mean()
print(f"Mean wickets: {mean_wickets}")

# Calculate the median
median_wickets = df['is_wicket'].median()
print(f"Median wickets: {median_wickets}")

# Calculate the mode
mode_wickets = stats.mode(df['is_wicket'])
print(f"Mode wickets: {mode_wickets}")
```
