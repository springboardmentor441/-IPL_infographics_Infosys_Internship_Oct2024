```python
import pandas as pd
deliveries_df = pd.read_csv("deliveries.csv")
player_stats = deliveries_df.groupby("batter").agg(
    Total_Runs=("batsman_runs", "sum"),
    Balls_Faced=("ball", "count"),
    Dismissals=("player_dismissed", "count")
).reset_index()
print(player_stats)

```

```python

player_stats["Batting Average"] = player_stats["Total_Runs"] / player_stats["Dismissals"]
player_stats["Batting Average"].fillna(0, inplace=True)


player_stats["Strike Rate"] = (player_stats["Total_Runs"] / player_stats["Balls_Faced"]) * 100
print(player_stats)

centuries = deliveries_df.groupby(["batter", "match_id"]).agg(Innings_Runs=("batsman_runs", "sum")).reset_index()
centuries_summary = centuries.groupby("batter").agg(
    Centuries=("Innings_Runs", lambda x: (x >= 100).sum()),
    Fifties=("Innings_Runs", lambda x: ((x >= 50) & (x < 100)).sum())
).reset_index()
print(centuries_summary)
#  Core Parameters:
#  Batting Average: [ {Batting Average} = {Total Runs Scored}/
#  {Number of Dismissals} ]
#  High averages indicate consistency.
#  Strike Rate: [ {Strike Rate} = ({{Total Runs Scored}}/
#  {{Balls Faced}}) *100 ]
#  Captures the batter's ability to score quickly.
#  Centuries (100's):
#  Count of individual innings where the batter scored 100 or more runs.
#  Fifties (50's):
#  Count of individual innings where the batter scored between 50 and 99
#  runs

```

```python

player_stats = player_stats.merge(centuries_summary, on="batter", how="left")
player_stats.fillna(0, inplace=True)  
print(player_stats)

metrics = ["Batting Average", "Strike Rate", "Centuries", "Fifties"]
normalized_metrics = {}
for metric in metrics:
    normalized_metrics[f"Normalized {metric}"] = (player_stats[metric] / player_stats[metric].max()) * 100
print(normalized_metrics)

for metric, normalized in normalized_metrics.items():
    player_stats[metric] = normalized
print(player_stats)

weights = {
    "Batting Average": 0.3,
    "Strike Rate": 0.4,
    "Centuries": 0.15,
    "Fifties": 0.05,
}


player_stats["Overall Score"] = (
    player_stats["Normalized Batting Average"] * weights["Batting Average"] +
    player_stats["Normalized Strike Rate"] * weights["Strike Rate"] +
    player_stats["Normalized Centuries"] * weights["Centuries"] +
    player_stats["Normalized Fifties"] * weights["Fifties"]
)
print(player_stats)
#assumed weights (initial weights)
# batting Average (30%): Measures consistency by dividing total runs by
#  dismissals.
#  Strike Rate (40%): Reflects the batter's ability to score quickly, critical in
#  T20 cricket.
#  Centuries (100's) (15%): Highlights match-winning capabilities.
#  Fifties (50's) (5%): Indicates the batter’s ability to contribute consistently
```

```python
# Rank players based on Overall Score
player_stats["Rank"] = player_stats["Overall Score"].rank(ascending=False)

top_players = player_stats.sort_values(by="Rank").head(10)
top_players[["batter", "Overall Score", "Rank"]]


 #Overall Score:
 #The Overall Score is computed as: [ {Overall Score} = summation of ({NormalizedMetric}*{Weight}) ]
 #Example:
 #If a player has:  Average: 55, Strike Rate: 140, Centuries: 4, Fifties: 6 And the best values in the dataset are:  Average: 60, Strike Rate: 150, Centuries:5, Fifties: 8
 #The normalized metrics will be: - Normalized Average = ( {55}/{60}* 100 =91.67 ) - Normalized Strike Rate = ( {140}/{150}*  100 = 93.33 ) Normalized Centuries = ( {4}/{5}* 100 = 80 ) - Normalized Fifties = ({6}/{8}* 100 = 75 )
 #Using weights, the Overall Score: [ {Overall Score} = (91.67 * 0.3) + (93.33* 0.4) + (80 * 0.15) + (75 * 0.05) = 88.33 ]
```
