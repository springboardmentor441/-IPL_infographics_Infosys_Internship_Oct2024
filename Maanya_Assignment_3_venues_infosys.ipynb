value_counts() method is a convenient way to count the number of occurrences of each unique value in a pandas Series or DataFrame column.

```python
venue_count=match_data.venue.value_counts()
venue_count
```

```python
top_venues=venue_count.head(15)
```

```python
top_cities=match_data.city.value_counts()
top_cities
```

```python
top_cities_15=top_cities.head(15)
```

```python
top_venues.plot(kind='bar',color='red')
plt.title('top venues')
plt.xlabel('venue name')
plt.ylabel('No.of matches')
```

```python
top_cities_15.plot(kind='bar',color='green')
plt.title('Top cities')
plt.xlabel('City')
plt.ylabel('No.of matches')
```

```python
venues_with_cities = match_data.groupby(['city', 'venue']).size().reset_index(name='Matches Count')
sorted=venues_with_cities.sort_values(by='Matches Count', ascending=False)
top_10_venues_with_city = sorted.head(15)
top_10_venues_with_city
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Matches Count', y='venue', hue='city', data=top_10_venues_with_city, palette='Set2') # hue adds extra information in the plot.
plt.title('Top 15 venues with city')
plt.xlabel('No.of Matches')
plt.ylabel('Venue')
plt.legend(title='City', loc='upper right')
plt.show()
print(top_10_venues_with_city)
```
