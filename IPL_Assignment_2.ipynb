```python
import pandas as pd
```

```python
df = pd.read_csv('deliveries.csv')
```

```python
df.info()
```

```python
batter_stats = df.groupby('batter').agg(
    total_runs=pd.NamedAgg(column='batsman_runs', aggfunc='sum'),
    total_balls=pd.NamedAgg(column='ball', aggfunc='count'),
    total_fours=pd.NamedAgg(column='batsman_runs', aggfunc=lambda x: (x == 4).sum()),
    total_sixes=pd.NamedAgg(column='batsman_runs', aggfunc=lambda x: (x == 6).sum()),
    total_wickets=pd.NamedAgg(column='is_wicket', aggfunc='sum')
).reset_index()
print(batter_stats)
```

```python
batter_stats['batting_average'] = batter_stats['total_runs'] / batter_stats['total_wickets'].replace(0, 1)  # Avoid division by zero
batter_stats['strike_rate'] = (batter_stats['total_runs'] / batter_stats['total_balls']) * 100
```

```python
# Sort players by total runs scored
batter_stats.sort_values(by='total_runs', ascending=False, inplace=True)
```

```python
#Display the top players
print(batter_stats[['batter', 'total_runs', 'batting_average', 'strike_rate', 'total_fours', 'total_sixes']])
```

```python
#task 1:calculate stats fpr runs scored by team and players
#runs by players
players_runs_stats={
    "Mean":df["batsman_runs"].mean(),
    "Median":df["batsman_runs"].median(),
    "Mode":df["batsman_runs"].mode()[0],
    "Standard Deviation":df["batsman_runs"].std(),
}
print("Player run stastatstics:")
print(players_runs_stats)
```

```python
import pandas as pd

team_runs_stats = df.groupby("batting_team")["total_runs"].agg(["mean", "median", lambda x: x.mode()[0], "std"])
# Using a lambda function to get the first mode value
print("Team run statistics:")
print(team_runs_stats)
```

```python
#task 2:calculate stats for wickets taken by bowlers
wickets_df=df[df["is_wicket"]==1]

wicket_stats={
    "Mean":wickets_df["is_wicket"].mean(),
    "Median":wickets_df["is_wicket"].median(),
    "Mode":wickets_df["is_wicket"].mode()[0],
}
print("\n Wicket statistics:")
print(wicket_stats)
```

```python
batter_stats=df.groupby('batter').agg(
    total_runs=('batsman_runs','sum'),
    times_out=('is_wicket','sum')
).reset_index()
batter_stats
```
