```python
import pandas as pd
```

```python
df=pd.read_csv('matches.csv')
df
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# Grouping the data by 'city' and counting the number of matches
city_match_count = df.groupby('city').size().reset_index(name='match_count')

# Sorting the cities by the number of matches in descending order
sorted_city_match_count = city_match_count.sort_values(by='match_count', ascending=False)

# Extracting the top 20 cities
top_20_cities = sorted_city_match_count.head(20)

# Displaying the result
print(top_20_cities)
```
