```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Load datasets
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

# Aggregate runs per season
season_runs = (
    deliveries.groupby('match_id')['total_runs']
    .sum()
    .reset_index()
    .merge(matches[['id', 'season']], left_on='match_id', right_on='id', how='left')
    .groupby('season')['total_runs']
    .sum()
    .reset_index()
)

# Set season as index
season_runs.set_index('season', inplace=True)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(season_runs, marker='o')
plt.title('Total Runs per Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.show()

# Check for stationarity
adf_test = adfuller(season_runs['total_runs'])
print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
if adf_test[1] < 0.05:
    print("Data is stationary.")
else:
    print("Data is not stationary. Differencing is needed.")

# Differencing if needed
season_runs_diff = season_runs.diff().dropna()

# Plot ACF and PACF
plot_acf(season_runs_diff)
plt.show()

plot_pacf(season_runs_diff)
plt.show()

# Fit ARIMA model
model = ARIMA(season_runs, order=(1, 1, 1))  # Adjust (p, d, q) based on analysis
arima_result = model.fit()

# Print model summary
print(arima_result.summary())

# Forecast future seasons
forecast = arima_result.forecast(steps=5)  # Predict for 5 future seasons
print("Forecasted Runs for Future Seasons:")
print(forecast)

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(season_runs, label='Actual')
plt.plot(range(season_runs.index[-1] + 1, season_runs.index[-1] + 6), forecast, label='Forecast', marker='o')
plt.title('ARIMA Model Forecast')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load your datasets (matches and deliveries)
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

# Calculate total runs scored and total balls faced by each batsman
batsman_stats = deliveries.groupby('batter').agg(
    total_runs=pd.NamedAgg(column='batsman_runs', aggfunc='sum'),
    total_balls=pd.NamedAgg(column='ball', aggfunc='count')
).reset_index()

# Calculate strike rate for each batsman
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['total_balls']) * 100

# Merge with matches dataset to get target runs for each match
match_stats = deliveries[['match_id', 'batting_team']].drop_duplicates()
match_stats = match_stats.merge(matches[['id', 'target_runs']], left_on='match_id', right_on='id', how='left')

# Calculate average target runs per match for each team
team_target_runs = match_stats.groupby('batting_team')['target_runs'].mean().reset_index()

# Merge team target runs with batsman statistics
batsman_stats = batsman_stats.merge(team_target_runs, left_on='batter', right_on='batting_team', how='left')

# Filter batsmen who have a minimum target run threshold (e.g., 150 runs)
min_target_runs = 150
filtered_batsman_stats = batsman_stats[batsman_stats['target_runs'] >= min_target_runs]

# Display filtered batsmen with their strike rates
print(filtered_batsman_stats[['batter', 'total_runs', 'total_balls', 'strike_rate', 'target_runs']])

# Optional: Plot the batsmen's strike rates
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.barh(filtered_batsman_stats['batter'], filtered_batsman_stats['strike_rate'], color='skyblue')
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.title(f'Strike Rate of Batsmen with Target Runs >= {min_target_runs}')
plt.show()
```

## Best all-rounder performance considering batting, bowling, and fielding

```python


# Step 1: Extracting batting, bowling, and fielding data

# Batting performance: Total runs scored by each batter
batting_performance = deliveries.groupby('batter')['batsman_runs'].sum()

# Bowling performance: Total wickets, total runs given, and balls bowled for each bowler
bowling_performance = deliveries.groupby('bowler').agg(
    wickets=('is_wicket', 'sum'),
    total_runs=('total_runs', 'sum'),
    balls_bowled=('ball', 'count')
)

# Calculate economy rate: Total runs given / overs bowled (balls / 6)
bowling_performance['economy_rate'] = bowling_performance['total_runs'] / (bowling_performance['balls_bowled'] / 6)

# Fielding performance: Dismissals made (how many players dismissed by the player)
fielding_performance = deliveries.groupby('fielder')['player_dismissed'].count()

# Step 2: Normalize the performance data
# For simplicity, let’s assume max values of runs, wickets, and dismissals to scale the data
max_runs = batting_performance.max()
max_wickets = bowling_performance['wickets'].max()
max_dismissals = fielding_performance.max()

batting_performance_points = (batting_performance / max_runs) * 100
bowling_performance_points = (bowling_performance['wickets'] / max_wickets) * 100
fielding_performance_points = (fielding_performance / max_dismissals) * 100

# Step 3: Combine the performance data into one DataFrame
performance_data = pd.DataFrame({
    'run_points': batting_performance_points,
    'wicket_points': bowling_performance_points,
    'dismissal_points': fielding_performance_points
})

# Step 4: Calculate the overall performance score (equal weights here, can be adjusted)
performance_data['overall_score'] = 0.33 * performance_data['run_points'] + 0.33 * performance_data['wicket_points'] + 0.33 * performance_data['dismissal_points']

# Step 5: Rank players based on the overall performance score and select top 10
top_10_performers = performance_data.sort_values(by='overall_score', ascending=False).head(10)

# Step 6: Output the top 10 all-rounders with their performance points and overall score
top_10_performers = top_10_performers.reset_index()
top_10_performers[['index', 'run_points', 'wicket_points', 'dismissal_points', 'overall_score']]
```

## . Win/Loss Analysis by Runs or Wickets for Each Team:


```python
# Merge matches and deliveries to analyze wins by runs or wickets
match_results = matches[['id', 'team1', 'team2', 'winner', 'result_margin', 'method']]

# Add columns for team1 and team2's win/loss results
match_results['team1_win_by_runs'] = match_results.apply(lambda x: 1 if x['winner'] == x['team1'] and x['result_margin'] > 0 else 0, axis=1)
match_results['team2_win_by_runs'] = match_results.apply(lambda x: 1 if x['winner'] == x['team2'] and x['result_margin'] > 0 else 0, axis=1)
match_results['team1_win_by_wickets'] = match_results.apply(lambda x: 1 if x['winner'] == x['team1'] and x['method'] == 'wickets' else 0, axis=1)
match_results['team2_win_by_wickets'] = match_results.apply(lambda x: 1 if x['winner'] == x['team2'] and x['method'] == 'wickets' else 0, axis=1)

# Summarize by each team - total wins by runs and by wickets
team_win_analysis = match_results.groupby(['team1']).agg(
    wins_by_runs=('team1_win_by_runs', 'sum'),
    wins_by_wickets=('team1_win_by_wickets', 'sum')
).reset_index()

# Display result
print(team_win_analysis)
```

## Head-to-Head Match Analysis Between IPL Teams:

```python
# Create a head-to-head matrix (win/loss) between teams
teams = matches['team1'].unique()

# Initialize an empty DataFrame to store the results
head_to_head_matrix = pd.DataFrame(0, index=teams, columns=teams)

# Fill the matrix with win/loss results
for _, row in match_results.iterrows():
    if row['winner'] == row['team1']:
        head_to_head_matrix.loc[row['team1'], row['team2']] += 1  # team1 wins
        head_to_head_matrix.loc[row['team2'], row['team1']] -= 1  # team2 loses
    elif row['winner'] == row['team2']:
        head_to_head_matrix.loc[row['team2'], row['team1']] += 1  # team2 wins
        head_to_head_matrix.loc[row['team1'], row['team2']] -= 1  # team1 loses

# Display the head-to-head matrix
print(head_to_head_matrix)
```
