```python
# Analyzing winner
winning=match_data['winner'].value_counts()
winning
```

```python
# Visualizing win
plt.figure(figsize=(14,7))
winning.plot(kind='bar',color='pink')
plt.title('Number of wins per team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
# Toss Decisions
toss=match_data['toss_decision'].value_counts()
toss
```

```python
plt.figure(figsize=(6,6))
plt.pie(toss, labels=toss.index, autopct='%1.1f%%', startangle=90)
plt.title("Toss Decision Distribution")
plt.show()
```

```python
# Winning of the team based on the TOSS decision
pd.crosstab(match_data['winner'],match_data['toss_decision']).plot(kind='bar',figsize=(12,6))
plt.title('Winning based on the TOSS decision')
plt.xlabel('Winning Team')
plt.ylabel('Count')
plt.show()
```
