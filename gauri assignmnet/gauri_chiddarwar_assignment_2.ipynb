### Compute the mean, median, and mode for wickets.

```python
# Filter for wicket events
wickets_data = df[df['is_wicket'] == 1]
```

```python
# Mean, median, and mode for wickets taken by players
wicket = wickets_data.groupby('bowler').size().reset_index(name='wickets')
```

```python
wicket['mean_wickets'] = wicket['wickets'].mean()
```

```python
wicket['median_wickets'] = wicket['wickets'].median()
```

```python
wicket['mode_wickets'] = wicket['wickets'].mode()[0]
```

### Third apporoch

```python
# -------------------------------------calculating the strike_rate ------------------------------------------------------
df['strike_rate'] = (df['batsman_runs'] / (df['ball'])) * 100
```

```python
# ----------------------------------------------- Calculate max values-----------------------------------------------------
max_average = player['mean_runs'].max()
max_strike = df['strike_rate'].max()
max_centuries = df.groupby('batter')['batsman_runs'].apply(lambda x: (x >= 100).sum()).max()
max_fifties = df.groupby('batter')['batsman_runs'].apply(lambda x: (x >= 50).sum()).max()
```

```python
#------------------------------------------------ Normalize---------------------------------------------------------------
player['normalized_average'] = (player['mean_runs'] / max_average) * 100
player['normalized_strike'] = (df['strike_rate'] / max_strike) * 100
player['normalized_centuries'] = (df.groupby('batter')['batsman_runs'].apply(lambda x: (x >= 100).sum()) / max_centuries) * 100
player['normalized_fifties'] = (df.groupby('batter')['batsman_runs'].apply(lambda x: (x >= 50).sum()) / max_fifties) * 100
```

Overall Score=(Normalized Avg×0.3)+(Normalized SR×0.4)+(Normalized Centuries×0.15)+(Normalized Fifties×0.05)

```python
player['overall_score'] = (
    player['normalized_average'] * 0.3 +
    player['normalized_strike'] * 0.4 +
    player['normalized_centuries'] * 0.15 +
    player['normalized_fifties'] * 0.05
)
```

```python
top_batsmen = player.sort_values(by='overall_score', ascending=False).head(5)
```
