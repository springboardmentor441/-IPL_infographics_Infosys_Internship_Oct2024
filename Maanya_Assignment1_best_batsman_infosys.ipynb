But you cannot judge these results as true because there can be the reason that one player played so many matches and scored less runs and total becomes more. And the other who played very well but played less number of matches.

So we need to find the best batsman using some statistical analysis.

Steps to Calculate the Best Batsman:

Load the Datasets: Read the matches.csv and deliveries.csv files.

Calculate Batting Average: For each batsman, calculate the average runs by dividing the total runs scored by the number of matches played.

Calculate Strike Rate: The strike rate is the number of runs scored per 100 balls faced.

Calculate Centuries and Fifties: Count how many times a batsman has scored 100 or 50 runs.

Normalize the Metrics: Normalize each metric (average, strike rate, centuries, fifties).

Calculate Overall Score: Use the weighted formula to calculate the overall score for each batsman.

Rank Players: Sort players based on their overall score in descending order.

```python
batsman_data=deliveries_data[deliveries_data['extras_type'].isna()]
```

```python
batsman_data
```

The reset_index method in Pandas resets the index of a DataFrame to the default one. After operations like filtering, concatenation, or merging, the index may no longer be sequential. This method helps re-establish a clean, sequential index.

```python
batsman_runs=batsman_data.groupby('batter')['total_runs'].sum().reset_index()
```

```python
batsman_runs
```

```python
matches_played_by_batsman=deliveries_data.groupby('batter')['match_id'].nunique().reset_index()
```

```python
matches_played_by_batsman.columns=['batter','matches']
```

```python
matches_played_by_batsman
```

```python
batsman_data_merged=pd.merge(batsman_runs,matches_played_by_batsman,on='batter',how='left')
```

```python
print(batsman_data_merged.head())
```

```python
# calculating batting average
batsman_data_merged['batting_average']=batsman_data_merged['total_runs']/batsman_data_merged['matches']
```

```python
batsman_data_merged
```

```python
# strike rate
balls_faced=batsman_data.groupby('batter')['ball'].count().reset_index()
batsman_data_merged=pd.merge(batsman_data_merged,balls_faced,on='batter',how='left')
```

```python
batsman_data_merged
```

```python
batsman_data_merged['strike_rate']=(batsman_data_merged['total_runs']/batsman_data_merged['ball'])*100
```

```python
batsman_data_merged
```

```python
# centuries
runs_per_match=deliveries_data.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
centuries=runs_per_match[runs_per_match['batsman_runs']>=100]
centuries_count=centuries.groupby('batter').size().reset_index(name='centuries')  # .size() to count the number of rows in each group. Each row represents a century scored by a batsman in a match.
```

```python
centuries_count
```

```python
batsman_data_merged = pd.merge(batsman_data_merged, centuries_count, on='batter', how='left')
batsman_data_merged
```

```python
batsman_data_merged[batsman_data_merged['batter'] == 'V Kohli']
```

There are some values in the centuries column which are null, so I am replacing them with 0.

```python
batsman_data_merged['centuries']=batsman_data_merged['centuries'].fillna(0).astype(int)
```

```python
batsman_data_merged
```

```python
# fifties
runs_per_match=deliveries_data.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
fifties=runs_per_match[(runs_per_match['batsman_runs'] >=50) & (runs_per_match['batsman_runs']<100)]
fifties_count=fifties.groupby('batter').size().reset_index(name='fifties')  # .size() to count the number of rows in each group. Each row represents fifties scored by a batsman in a match.
```

```python
fifties_count
```

```python
batsman_data_merged = pd.merge(batsman_data_merged, fifties_count, on='batter', how='left')
batsman_data_merged
```

```python
batsman_data_merged[batsman_data_merged['batter'] == 'V Kohli']
```

```python
batsman_data_merged['fifties']=batsman_data_merged['fifties'].fillna(0).astype(int)
```

```python
batsman_data_merged
```

```python
# Normalize metrics
batsman_data_merged['normalized_average'] = batsman_data_merged['batting_average'] / batsman_data_merged['batting_average'].max()
batsman_data_merged['normalized_strike_rate'] = batsman_data_merged['strike_rate'] / batsman_data_merged['strike_rate'].max()
batsman_data_merged['normalized_centuries'] = batsman_data_merged['centuries'] / batsman_data_merged['centuries'].max()
batsman_data_merged['normalized_fifties'] = batsman_data_merged['fifties'] / batsman_data_merged['fifties'].max()
```

```python
batsman_data_merged
```

```python
batsman_data_merged['normalized_centuries'] = batsman_data_merged['normalized_centuries'].fillna(0)
batsman_data_merged['normalized_fifties'] = batsman_data_merged['normalized_fifties'].fillna(0)
```

```python
batsman_data_merged
```

```python
# assign weights
w1, w2, w3, w4 = 0.3, 0.4, 0.15, 0.05
```

```python
batsman_data_merged['overall_score'] = (
    batsman_data_merged['normalized_average'] * w1 +
    batsman_data_merged['normalized_strike_rate'] * w2 +
    batsman_data_merged['normalized_centuries'] * w3 +
    batsman_data_merged['normalized_fifties'] * w4
)
```

```python
batsman_data_merged
```

```python
best_batsman = batsman_data_merged.sort_values('overall_score', ascending=False)
```

```python
print(best_batsman[['batter', 'overall_score']])
```
