```python
batter_match_runs=df.groupby(["match_id","batter"])["batsman_runs"].sum().reset_index()
print(batter_match_runs)

batter_stats=batter_match_runs.groupby("batter")["batsman_runs"].sum().reset_index()
batter_stats.columns=["batter","total_run"]
batter_stats
```

```python
#The code calculates batting stats by summing dismissals (is_wicket), counting unique matches played (match_id), and merging these with total runs
#into a single DataFrame, providing batter, total_runs, total_times_out, and total_matches_played for each player.
```

```python
batter_knockout=df.groupby("batter")["is_wicket"].sum()
print(batter_knockout)
batter_info=pd.merge(batter_stats,batter_knockout,on="batter",how="left")
batter_info.columns=["batter","total_runs","total_times_out"]

total_matches_played=df.groupby("batter")["match_id"].nunique().reset_index()

total_matches_played.columns=["batter", "total_matches_played"] 
batter_stats=pd.merge(batter_info,total_matches_played, on="batter", how="left")
```

```python
#The code calculates each batter's batting average as total_runs / total_times_out, replacing infinite values (from division by zero) with NaN. 
#It also sets the batting average to NaN for batters who were never out (total_times_out = 0). Finally, it sorts the DataFrame by batting average in 
#descending order and selects the top 5 batters with the highest averages.
```

```python

batter_stats["batting_average"]=batter_stats["total_runs"]/batter_stats["total_times_out"]
batter_stats["batting_average"].replace(np.inf,np.nan,inplace=True)
batter_stats.loc[batter_stats["total_times_out"]==0,"batting_average"]=np.nan
top_5_batting_avgerage=batter_stats.sort_values(by="batting_average",ascending=False).head(5)
top_5_batting_avgerage.head(5)
```

```python
#The code calculates a custom player_score by combining the batter's batting_average, a scaled contribution from their total_runs (divided by 1000),
#and a scaled contribution from total_matches_played (divided by 10). It then sorts the DataFrame in descending order of player_score to identify the 
#best-performing players and selects the top 5 players with the highest scores.
```

```python
batter_stats["player_score"]=(batter_stats["batting_average"]+(batter_stats["total_runs"]/1000)+(batter_stats["total_matches_played"]/10))
best_player=batter_stats.sort_values(by="player_score",ascending=False)
best_player.head(5)
```
