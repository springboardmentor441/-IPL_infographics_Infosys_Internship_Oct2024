## Grouping and Sorting

```python
import pandas as pd
```

```python
Matches = pd.read_csv("matches.csv")
```

```python
Matches.groupby('result_margin').result_margin.count()
```

```python
Matches.groupby('result_margin').target_runs.min()
```

```python
Matches.groupby('venue').apply(lambda df: df.team1.iloc[0])
```

```python
Matches.groupby(['city', 'season']).apply(lambda df: df.loc[df.result_margin.idxmax()])
```

```python
Matches.groupby('team1').target_runs.agg([len, min, max])
```

## Multi-indexes

```python
teams_venues_reviewed = Matches.groupby(['team1', 'venue']).winner.agg([len])
teams_venues_reviewed
```

```python
mi = teams_venues_reviewed.index
type(mi)
```

```python
teams_venues_reviewed_reset = teams_venues_reviewed.reset_index()
teams_venues_reviewed_reset
```

## Sorting

```python
# Group by 'team1' and 'venue', and get the count of 'winner'
teams_venues_reviewed = Matches.groupby(['team1', 'venue']).winner.agg([len])

# Reset the index
teams_venues_reviewed = teams_venues_reviewed.reset_index()
teams_venues_reviewed
```

```python
teams_venues_reviewed.sort_values(by='len')
```

```python
teams_venues_reviewed.sort_values(by='len', ascending=False)
```

```python
teams_venues_reviewed.sort_index()
```

```python
teams_venues_reviewed.sort_values(by=['team1', 'len'])
```
