# **Design Document: Analysis of Top 20 Cities with the Most Matches**

## **Objective**
To identify and analyze the top 20 cities where the highest number of matches were played in the dataset.

---

## **Data Input**
- **Dataset Columns**:  
  - `id`, `season`, `city`, `date`, `match_type`, `player_of_match`,  
    `venue`, `team1`, `team2`, `toss_winner`, `toss_decision`, `winner`,  
    `result`, `result_margin`, `target_runs`, `target_overs`, `super_over`,  
    `umpire1`, `umpire2`.  
- **Key Column for Analysis**:  
  - `city`

---

## **Processing Steps**
1. **Extract City Information**:  
   Retrieve the count of matches played in each city using the `value_counts()` method.

2. **Select Top 20 Cities**:  
   Sort the counts in descending order and select the top 20 entries.

3. **Prepare the Data for Presentation**:  
   Convert the results into a tabular format with the columns:
   - `City`: Name of the city.
   - `Number of Matches`: Count of matches played in that city.
