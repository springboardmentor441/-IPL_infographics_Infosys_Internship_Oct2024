```
import pandas as pd
```

```
matches=pd.read_csv('matches.csv')
```

```
matches
```

```
filtered_data=matches[['city','venue']]
filtered_data
```

```
matches_city=filtered_data.groupby(['city','venue']).size().reset_index(name='number_of_matches_played')
matches_city
```

```
particular_city=filtered_data[filtered_data['city']=='Chennai'].groupby(['city','venue']).size().reset_index(name='number_of_matches_played')
particular_city
```

```
top_20_cities = matches['city'].value_counts().nlargest(20).reset_index(name='number_of_matches_played')
top_20_cities
```
